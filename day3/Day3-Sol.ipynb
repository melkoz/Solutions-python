# a)

```python
class Person(object):
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        
    def print_person(self):
        print(self.firstname+' '+self.lastname)
```

```python
someone = Person('George', 'Sand')
someone.print_person()
```

# b)

```python
class Student(Person):
    def __init__(self, firstname, lastname, subject): 
        Person.__init__(self,firstname, lastname ) 
        self.subject = subject
    def print_student(self):
        print(self.firstname+' '+self.lastname +'\n'+'Subject area: ' + self.subject)
        
```

```python
some_student = Student('Rosalind', 'Franklin', 'Structure of DNA')
some_student.print_student()
```

# c)

```python
class Teacher(Person):
    def __init__(self, firstname, lastname, course): 
        Person.__init__(self,firstname, lastname ) 
        self.course = course
    def print_teacher(self):
        print(self.firstname+' '+self.lastname +'\n'+'Course of the teacher: ' + self.course)
        
```

```python
some_teacher = Teacher('Filipe', 'Maia', 'Python programming')
some_teacher.print_teacher()
```

```python
# Program to multiply two matrices using nested loops
import random

n = 250


def x_gen(N):
    # NxN matrix
    X = np.random.randint(0,100,(N,N))
    return X


def y_gen(N):
    # Nx(N+1) matrix
    Y = np.random.randint(0,100,(N,N+1))
    return Y


def result_gen(N):
    # result is Nx(N+1)
    result = np.zeros((N,N+1))
    return result


def multipl(X,Y,result):

    result=  np.matmul(X,Y)
    return result 


def printing(result):
    for r in result:
        print(r)

x = x_gen(n)
y = y_gen(n)
res = result_gen(n)

result_final = multipl(x, y, res)
printing(result_final)
```
